### Creating groupby for GEO_Region and Passenger_Count to find which have largest and smallest passenger count 

```python
k=df.groupby('GEO_Region')['Passenger_Count'].sum().nlargest(5)
```

```python
sb.barplot(x=k.index,y=k.values,palette='copper')
plt.title("GEO_Region and  Passenger_Count distribution")
plt.xlabel("GEO_Region")
plt.ylabel("Passenger_Count")
plt.show()
```

```python
s=df.groupby('GEO_Region')['Passenger_Count'].sum().nsmallest(5)
```

```python
plt.figure(figsize=(15,5))
sb.barplot(x=s.index,y=s.values,palette='copper')
plt.show()
```

###  1.South america has the lowest passenger count as shown in above graph.
### 2.Mexico has the highest passenger count.
### 3.Central America , Middle East and Oceania has the moderate passenger count.

```python
z=pd.pivot_table(df,index="Year",columns="Month",values="Passenger_Count",aggfunc="sum")
z
```

```python
plt.figure(figsize=(18,8))
sb.lineplot(x=df.Month,y=df.Passenger_Count)
plt.title("sesonal trends of passenger traffic")
plt.xlabel("month")
plt.ylabel("Passenger count")
plt.show()

```

### 1.The graph shows march month is growing its passenger traffic as it is the pre-summer month.
### 2.February indicating the least passenger traffic that shows february is the least busy month for travel.
### 3.Traffic is steadily arising from the month of june .
### 4.July ,August and september is showing the lowest of lowest passenger count indicates traffic of passengers is or equal to zero.

# What trends can be changed from year to year passenger traffic ?

```python
plt.figure(figsize=(15,5))
sb.heatmap(z,annot=True)
plt.title("monthly passenger traffic of years ")
plt.xlabel("month")
plt.ylabel("year")
plt.show()
```

# Terminal and Boarding Area Analysis:

## Which terminals are busy showing the most and least passenger traffic?

```python
a=df.groupby(['Terminal','Boarding_Area'])['Passenger_Count'].sum().sort_values().reset_index()
a
            
```

```python
sb.barplot(x=a.Boarding_Area,y=a.Passenger_Count,hue=a.Terminal,palette='copper')
plt.legend(title='terminal',bbox_to_anchor=(1.0,1.0))
plt.show()
```

### 1.Most passengers travelled through terminal 3 that is area F,showing it the busiest terminal.
### 2.A and G shows the international terminal ,it shows many people travelled from foreign countries .
### 3.Terminal 1 and 2 shows the traffic for domestic airlines .
